## Creando los agentes

```python
from crewai import Crew, Process, Agent, Task


# Define tus agentes
desenvolvedor_de_ideais = Agent(
  #Papel: Define lo que el agente hace y su área de especialización.
  role='Creador de Ideas Creativas para Publicaciones de Blog', #IDENTIDAD
  #Objetivo: La finalidad y motivación del agente
  goal='Generar ideas originales y relevantes para publicaciones de blog dentro de un tema específico',
  #Histórico: Experiencia y perspectiva del agente
  backstory='Eres un creador creativo, siempre al tanto de las últimas tendencias y con una habilidad impresionante para transformar conceptos en ideas innovadoras. Tu curiosidad y energía te ayudan a crear sugerencias únicas que atraen eficazmente al público objetivo.',
  verbose=True
)

planejador_de_conteudo = Agent(
  role='Estratega de Contenido para Blogs', #IDENTIDAD
  goal='Planificar y estructurar el contenido de manera eficaz, basándose en el briefing proporcionado',
  backstory='Eres un estratega detallista, apasionado por alinear objetivos y datos con la creación de contenido. Te encanta crear planes bien estructurados que guíen a los redactores para alcanzar los mejores resultados. Tu enfoque está siempre en asegurar que el contenido cumpla con las expectativas del público y los objetivos de marketing.',
  verbose=True
)

escritor_do_post = Agent(
  role='Redactor Creativo de Publicaciones de Blog', #IDENTIDAD
  goal='Escribir publicaciones de blog atractivas y de alta calidad, siguiendo el briefing y las directrices definidas',
  backstory='Eres un escritor versátil, capaz de adaptar tu estilo de escritura al tono y formato deseado. Tu objetivo es siempre crear contenido claro, interesante y que capte la atención del lector, transformando ideas e información en historias atractivas y bien estructuradas.',
  verbose=True)

revisor_do_post = Agent(
  role='Revisor de Contenido de Publicaciones de Blog', #IDENTIDAD
  goal='Asegurar que la publicación esté libre de errores y lista para ser publicada',
  backstory='Eres un revisor minucioso y atento a los detalles. Tu misión es corregir errores ortográficos, mejorar la fluidez del texto y asegurar que el contenido esté perfectamente alineado con el briefing y los estándares de calidad. Tu trabajo es garantizar que cada publicación esté impecable antes de su publicación.',
  verbose=True)
```

## Creando las Tasks

```python
# Define tus tareas
cria_ideias = Task(
  #Descripción de la Tarea: El proceso
  description='Crea una lista con 10 ideas diferentes para publicaciones de blog sobre el tema: {tema}. Las ideas deben ser creativas, relevantes y diversas en formato, con foco en atraer al público objetivo.',
  agent=desenvolvedor_de_ideais,
  #Salida Esperada de la Tarea: El entregable. Esto es como el Output Parser de LangChain.
  expected_output='Una lista con 10 ideas de publicaciones, cada una con un título creativo y un breve resumen del contenido que se abordará.'
)

seleciona_ideias = Task(
  description='Selecciona la mejor idea de la lista generada, justificando tu elección con base en la relevancia y el alineamiento con los objetivos del blog.',
  agent=desenvolvedor_de_ideais,
  expected_output='La elección de una idea con una justificación clara y concisa sobre su relevancia y alineamiento con los objetivos del contenido.'
)

planeja_conteudo = Task(
  description='Crea un briefing detallado para la publicación del blog, incluyendo información como objetivo, público objetivo, tono de voz, palabras clave y formato.',
  agent=planejador_de_conteudo,
  expected_output='Un briefing estructurado, abordando todos los puntos importantes para guiar la creación del contenido.'
)

escreve_post = Task(
  description='Escribe el contenido completo de la publicación del blog, siguiendo las directrices del briefing y la idea seleccionada. Asegúrate de que la publicación sea atractiva, bien estructurada y adecuada al público objetivo.',
  agent=escritor_do_post,
  expected_output='Una publicación de blog con introducción, desarrollo y conclusión, que sea clara, interesante y alineada con el briefing.'
)

revisa_post = Task(
  description='Revisa la publicación del blog, corrigiendo errores gramaticales y de puntuación, además de mejorar la fluidez del texto. Asegúrate de que la publicación esté alineada con el tono y objetivo definidos en el briefing.',
  agent=revisor_do_post,
  expected_output='La publicación revisada, sin errores gramaticales y con buena fluidez, lista para ser publicada.'
)
```

## Creando la Crew

```python
blog_post_creation_crew = Crew(
  agents=[desenvolvedor_de_ideais, planejador_de_conteudo, escritor_do_post, revisor_do_post],
  tasks=[cria_ideias, seleciona_ideias, planeja_conteudo, escreve_post, revisa_post],
  process=Process.sequential #Si lo comparas con LangGraph vendría siendo similar al StatefulGraph.
                #hierarquical, este vendría siendo más parecido a la arquitectura tipo Supervisor (create_react_agent).
)
```

## Ejecutando la Crew

```python
result = await blog_post_creation_crew.kickoff_async({'tema': 'Realizando aplicaciones IA con Claude Code'})
print(result.raw)
```
